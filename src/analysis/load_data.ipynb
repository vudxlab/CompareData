```python
import pandas as pd
from pathlib import Path

# Đọc các file trong thư mục data/processed
processed_dir = Path('../../data/processed')
csv_files = list(processed_dir.glob('*.csv'))

print(f"Tìm thấy {len(csv_files)} file CSV trong thư mục processed:")
for file in csv_files:
    print(f"  - {file.name}")

print("\n" + "="*60)
print("10 HÀNG ĐẦU CỦA TỪNG FILE:")
print("="*60)

# Đọc và hiển thị 10 hàng đầu của mỗi file
for file in csv_files:
    print(f"\n📁 File: {file.name}")
    print("-" * 50)
    try:
        df = pd.read_csv(file)
        print(f"Kích thước: {df.shape[0]} hàng × {df.shape[1]} cột")
        print(f"Các cột: {list(df.columns)}")
        print("\n10 hàng đầu:")
        print(df.head(10).to_string())
    except Exception as e:
        print(f"Lỗi khi đọc file: {e}")
    print("\n" + "="*60)
```
